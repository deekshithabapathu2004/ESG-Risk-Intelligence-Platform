```python
# =============================================
# STEP 1: Load and Validate Raw ESG Dataset
# =============================================
import pandas as pd

# Load raw data (no header)
df = pd.read_csv("data/company_esg_financial_dataset.csv", header=None)

# Assign CORRECT column names (based on your data structure)
columns = [
    "company_id",
    "company_name",
    "industry",
    "region",
    "year",
    "revenue",
    "profit_margin",
    "market_cap",
    "growth_rate",
    "esg_overall",
    "esg_environmental",
    "esg_social",
    "esg_governance",          # ✅ This is the correct name
    "carbon_emissions",
    "water_usage",
    "energy_consumption"
]

df.columns = columns

# Convert to numeric (safe)
numeric_cols = [
    "year", "revenue", "profit_margin", "market_cap", "growth_rate",
    "esg_overall", "esg_environmental", "esg_social", "esg_governance",
    "carbon_emissions", "water_usage", "energy_consumption"
]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Show first 5 rows
print("✅ Data loaded successfully!")
print("\nFirst 5 rows:")
df.head()
```
